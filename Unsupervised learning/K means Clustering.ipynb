# K means Clustering implementation

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs

X , y = make_blobs(n_samples=1000 , centers=3 , random_state=1 , n_features=2)
```

```python
# we will not not be using y
X
```

```python
plt.scatter(X[:,0] , X[:,1] , c=y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X , y , test_size = 0.3 , random_state = 0)
```

```python
X_train.shape ,  y_test.shape
```

```python
from sklearn.cluster import KMeans

KMean_cluster = KMeans(n_clusters=3)

KMean_cluster
```

```python
KMean_cluster.fit(X_train)
```

```python
y_labels = KMean_cluster.predict(X_train)
y_labels
```

```python
# old data --> X and y

plt.scatter(X[:,0] , X[:,1] , c=y)
plt.show()
```

```python
# on train data , y_label is predicted on train data

plt.scatter(X_train[:,0] , X_train[:,1] , c=y_labels)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_score(X_train , y_labels)*100
```

```python
KMean_cluster.labels_
```

```python
KMean_cluster.inertia_
```

```python
# how to selected K(centroid) >>>  elbow method or silhouette method

wcss = []

for i in range(1 , 10):
    kmean = KMeans(n_clusters=i)
    kmean.fit(X_train)
    kmean.predict(X_train)
    wcss.append(kmean.inertia_)


print(wcss)
```

```python
plt.plot(range(1,10) , wcss)
plt.xticks(range(1,10))
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
