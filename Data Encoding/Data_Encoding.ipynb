# Nominal /  One-Hot Encoding

```python
pip install -U scikit-learn
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({'Status' : ['single' , 'married' , 'separated', 'single' , 'married' , 'separated' , 'separated']})

df
```

```python
encoder = OneHotEncoder()
encoder
```

```python
df_encoder = encoder.fit_transform(df[['Status']]).toarray()
```

```python
encoder.get_feature_names_out()
```

```python
encoded_data = pd.DataFrame(df_encoder , columns = encoder.get_feature_names_out())
```

```python
encoder.transform([['single']]).toarray()
```

# Practical Implementation in Datasets

```python
pd.concat([df , encoded_data] , axis=1)
```

```python
import seaborn as sns

data = sns.load_dataset("tips")
```

```python
encoders = OneHotEncoder()
encodered = encoders.fit_transform(data[['sex' , 'smoker' , 'time']]).toarray()
```

```python
df_encoded = pd.DataFrame(encodered , columns = encoders.get_feature_names_out())
```

```python
pd.concat([data , df_encoded] , axis=1)
```

# Ordinal Encoder

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.DataFrame({'class' : ['HS' , 'INTER' , 'GR' , 'PG' , 'PhD ', 'INTER' , 'GR' , 'PG' , 'PhD ']})


df

```

```python
encoder = OrdinalEncoder(categories=[['HS' , 'INTER' , 'GR' , 'PG' , 'PhD ']])
encoder.fit_transform(df[['class']])
```
