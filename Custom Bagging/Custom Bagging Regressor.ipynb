# Custom Bagging Regressor

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

***make_regression***

```python
from sklearn.datasets import make_regression

# Generate dataset
X, y = make_regression(n_samples=1000, n_features=2, noise=10, random_state=1)
```

***train_test_split***


```python
from sklearn.model_selection import train_test_split


# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
```

***Models***

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
LinearRegression_model = LinearRegression()

DecisionTreeRegressor_model = DecisionTreeRegressor()

SVR_model = SVR()   
```

# Ensemble 

***from sklearn.ensemble import VotingRegressor***

```python
from sklearn.ensemble import VotingRegressor

ensembles = VotingRegressor(estimators = [('LinearRegression' , LinearRegression_model) , ('DecisionTreeRegressor' , DecisionTreeRegressor_model) , ('SVR_model' , SVR_model)] , weights=None)
ensembles
```

```python
ensembles.fit(X_train , y_train)
```

```python
y_pred = ensembles.predict(X_test)
y_pred
```

***r2_score***

```python
from sklearn.metrics import r2_score

r2_score(y_test , y_pred)*100
```
