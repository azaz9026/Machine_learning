# Ensemble Learning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset('iris')
data
```

```python
data.species.unique()
```

```python
df = data['species']
```

```python
from sklearn.preprocessing import LabelEncoder

encder = LabelEncoder()

data['encder_species'] = encder.fit_transform(df)
```

```python
data.encder_species.unique()
```

```python
data.head(3)
```

```python
data.drop(columns=['species'] , axis=1 ,  inplace=True)
```

```python
data.head(2)
```

```python
X = data.iloc[:: ,:-1]
y = data.iloc[:,-1]
```

```python
X , y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X , y , test_size = 0.3 , random_state = 0)
```

```python

X_train.shape , y_train.shape
```

```python

X_test.shape , y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC 
from sklearn.tree import DecisionTreeClassifier
```

# Linear Regression

```python
from sklearn.model_selection import GridSearchCV

param_grid_lr = {
    'fit_intercept' :  [True , False],
    'copy_X' : [True , False],
    'n_jobs' : [1 , 2 , 3],
    'positive' : [True , False]
}

gs = GridSearchCV(LinearRegression() , param_grid_lr , cv = 5)
gs.fit(X_train, y_train)
```

```python
gs.best_params_
```

```python
lr = LinearRegression(copy_X = True , fit_intercept = False , n_jobs = 1 , positive = False)
lr.fit(X_train, y_train) 
```

```python
y_pred = lr.predict(X_test)
```

```python
lr.score(X_train, y_train)*100  ,  lr.score(X_test, y_test)*100
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import GridSearchCV

param_grid_lgr = {
    'penalty': ['l1', 'l2', 'elasticnet', 'none'],
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
    'max_iter': [100, 200, 300]
}

gs = GridSearchCV(LogisticRegression() , param_grid_lgr , cv = 5)
gs.fit(X_train, y_train)
```

```python
gs.best_params_
```

```python
LogisticR = LogisticRegression(C=1 , max_iter=100 , penalty='l1' , solver='saga')
LogisticR.fit(X_train, y_train)
```

```python
LogisticR.score(X_train, y_train)*100 ,  LogisticR.score(X_test, y_test)*100
```

# Support Vector Machine (SVM)

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import GridSearchCV

param_grid_svc = {
    'C': [0.1, 1, 10, 100, 1000],
    'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid']
}

gs = GridSearchCV(SVC() , param_grid_svc , cv = 5)
gs.fit(X_train, y_train)
```

```python
gs.best_params_
```

```python
svc = SVC(C=10 , gamma=0.1 , kernel='rbf')
svc.fit(X_train, y_train)
```

```python
svc.score(X_train, y_train)*100 ,  svc.score(X_test, y_test)*100
```

# Decision Tree Classifier

```python
from sklearn.model_selection import GridSearchCV

param_grid_tree = {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random'],
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': [None, 'sqrt', 'log2']
}

gs = GridSearchCV(DecisionTreeClassifier() , param_grid_tree, cv=5, n_jobs=-1)
gs.fit(X_train, y_train)
```

```python
gs.best_params_
```

```python
tree = DecisionTreeClassifier(criterion='gini' , max_depth=10 , max_features= None , min_samples_leaf=4 , min_samples_split=10 , splitter='random')
tree.fit(X_train, y_train)
```

```python
tree.score(X_train, y_train)*100 ,  tree.score(X_test, y_test)*100
```

```python
from sklearn.ensemble import VotingClassifier

models = [
    ('logistic_regression', LogisticRegression()),
    ('svc', SVC()),
    ('decision_tree', DecisionTreeClassifier())
]
```

```python
# Create the VotingClassifier
voting_c = VotingClassifier(estimators=models, voting='soft')

# Fit the VotingClassifier to the training data
voting_c.fit(X_train, y_train)
```

```python

voting_c.score(X_train, y_train)*100 , voting_c.score(X_test, y_test)*100 # Calculate training and testing scores
train_score = voting_c.score(X_train, y_train) * 100
test_score = voting_c.score(X_test, y_test) * 100

print(f"Training Accuracy: {train_score:.2f}%")
print(f"Testing Accuracy: {test_score:.2f}%")
```
