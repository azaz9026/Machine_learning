# Ordinal Encoder

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('titanic')
```

```python
df
```

```python
df['class'].unique()
```

```python
df['class'].value_counts()
```

```python
ord_class = [['First' , 'Second' , 'Third']]
ord_class
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ord = OrdinalEncoder(categories=ord_class)

ord.fit(df[['class']])
df['Encoded_class']  =  ord.transform(df[['class']])
```

```python
df['Encoded_class'].unique()
```

```python
df
```

***Ordinal Encoding by map function***

```python
ord_data = {
    'First': 1,
    'Second': 2,
    'Third': 3
}
```

```python
df['data_class_encoded'] = df['class'].map(ord_data)
```

```python
df['data_class_encoded'].unique()
```
